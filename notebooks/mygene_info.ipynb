```python
import mygene
import json
from rich import print as rprint

mg = mygene.MyGeneInfo()

def fetch_and_display_gene_data(gene_id):
    query_results = mg.query(gene_id, fields='all')
    if query_results['total'] == 0:
        print(f"No results found for {gene_id}")
        return
    gene_data = query_results['hits'][0] #get the first result only
    print(f"\n--- Results for: {gene_id} ---")
    rprint(json.dumps(gene_data, indent=4)) # using json to print data more cleanly.

    # Example of accessing nested data
    print("\n---Example Access of Nested Fields----")
    print("  Symbol:", gene_data.get('symbol', 'Not Available'))
    print("  Name:", gene_data.get('name', 'Not Available'))
    print("  Entrez ID:", gene_data.get('entrezgene', 'Not Available'))

    homologs = gene_data.get('agr', {}).get('orthologs', []) # accessing nested homologs
    if homologs:
        print("  Homologs (first 5):")
        for h in homologs[0:5]:
            print(f"    {h.get('symbol', 'Not Available')} ({h.get('taxid', 'N/A')})")
    else:
       print("  No homologs found")



# Example Usage
gene_ids = ['0610009B22Rik', 'Trappc2b', '58324']  # Add more gene IDs if you want

for gene_id in gene_ids:
    fetch_and_display_gene_data(gene_id)
```
