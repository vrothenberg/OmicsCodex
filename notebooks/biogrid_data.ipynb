```python
import pandas as pd


def preprocess_biogrid_df(df: pd.DataFrame) -> pd.DataFrame:
    """Preprocesses the DataFrame for faster gene searches.

    Args:
        df (pd.DataFrame): The original BioGRID DataFrame.

    Returns:
        pd.DataFrame: A new DataFrame with a preprocessed search column.
    """
    # Convert to uppercase
    df[[
        "Systematic Name Interactor A",
        "Systematic Name Interactor B",
        "Official Symbol Interactor A",
        "Official Symbol Interactor B",
    ]] = df[[
        "Systematic Name Interactor A",
        "Systematic Name Interactor B",
        "Official Symbol Interactor A",
        "Official Symbol Interactor B",
    ]].apply(lambda x: x.str.upper())

    # Handle NaN values in synonym columns
    for col in ["Synonyms Interactor A", "Synonyms Interactor B"]:
        df[col] = df[col].fillna("")

    # Split synonyms into lists
    df["Synonyms Interactor A"] = df["Synonyms Interactor A"].str.upper().str.split("|")
    df["Synonyms Interactor B"] = df["Synonyms Interactor B"].str.upper().str.split("|")

    # Create a combined search column
    df["searchable_text"] = (
        df["Systematic Name Interactor A"].fillna("") + " "
        + df["Systematic Name Interactor B"].fillna("") + " "
        + df["Official Symbol Interactor A"].fillna("") + " "
        + df["Official Symbol Interactor B"].fillna("") + " "
        + df["Synonyms Interactor A"].apply(lambda x: " ".join(x)) + " "
        + df["Synonyms Interactor B"].apply(lambda x: " ".join(x))
    ).str.strip()  # Remove leading/trailing whitespace

    df = df.set_index("searchable_text")

    return df


def get_gene_interactions(gene: str, df: pd.DataFrame) -> pd.DataFrame:
    """Searches for a given gene in a preprocessed DataFrame.

    Args:
        gene (str): The gene name to search for.
        df (pd.DataFrame): The preprocessed DataFrame.

    Returns:
        pd.DataFrame: A DataFrame containing rows where the gene was found.
    """
    gene = gene.upper()

    # Split the gene into a list
    if " " in gene:
        list_of_genes = gene.split(" ")
    else:
        list_of_genes = [gene]

    # Search using the index
    try:
        tmp_df = df.loc[df.index.isin([x for x in df.index if any(target in x for target in list_of_genes)])]
    except KeyError:
        return pd.DataFrame()

    return tmp_df.reset_index(drop=False)



df = pd.read_csv("/Users/vince/Salk/OmicsCodex/data/biogrid/biogrid_df.csv")

# Preprocess the DataFrame
processed_df = preprocess_biogrid_df(df)
```
