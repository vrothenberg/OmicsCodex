```python
import asyncio
from pyppeteer import launch
import logging

# Basic logging setup for jupyter
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')


async def fetch_html_test(url):
    """Fetches HTML using Pyppeteer and prints it."""
    browser = None
    try:
        browser = await launch(headless=True, args=['--no-sandbox', '--disable-setuid-sandbox'])
        page = await browser.newPage()
        await page.setViewport({'width': 1920, 'height': 1080})
        await page.goto(url, {'waitUntil': 'domcontentloaded'})
        await page.waitForSelector("div#References", {'timeout': 20000})
        html_content = await page.content()
        logging.info(f"Successfully fetched HTML from {url}")
        print(html_content[:500]) # Print a snippet of the HTML
        return html_content
    except Exception as e:
      logging.error(f"Error during fetch: {e}")
      return None
    finally:
        if browser:
            await browser.close()


# Example URL to test with
test_url = "https://wikicrow.ai/CLOCK"

# Run the async function in the notebook event loop
html = await fetch_html_test(test_url)

if html:
    print("HTML fetched successfully.")
else:
    print("Failed to fetch HTML.")
```
